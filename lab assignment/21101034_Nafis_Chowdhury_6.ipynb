```python
f1=open("input_file.txt","r")
p={}
graph={}
h={}
d={}
dh={}
length={}
import math
n=len(f1.readlines())
f1=open("input_file.txt","r")
for i in range(n):
    arr=f1.readline().split(" ")
    m=len(arr)

    x=arr[0].strip("\n")
    y=int(arr[1].strip("\n"))
    h[x]=y
    graph[x]=[]
    if "\n" in arr:
        for j in range(2,m-1,2):

            graph[x]=graph[x]+[arr[j]]
            if arr[j] not in graph.keys():
                graph[arr[j]]=[x]
            else:
                if arr[j] not in graph.keys():
                    graph[arr[j]]=graph[arr[j]]+[x]
            length[x,arr[j]]=int(arr[j+1])
    else:
        for j in range(2,m,2):

            graph[x]=graph[x]+[arr[j]]
            if arr[j] not in graph.keys():
                graph[arr[j]]=[x]
            else:
                if arr[j] not in graph.keys():
                    graph[arr[j]]=graph[arr[j]]+[x]
            length[x,arr[j]]=int(arr[j+1])




def priority_queue(pq):
    n=len(pq)

    for i in range(n-1):
        for j in range(n-i-1):
            if pq[j][1]>pq[j+1][1]:
                pq[j], pq[j+1] = pq[j+1], pq[j]

    a=pq.pop(0)
    return a


def search(graph,s,g,flag):
    pq=[]
    for i in graph.keys():
        d[i]=math.inf
        p[i]=None
        dh[i]=math.inf
    d[s]=0
    dh[s]=h[s]
    pq.append([s,dh[s]])

    while len(pq)>0:
        a=priority_queue(pq)
        u=a[0]

        for v in graph[u]:
            if d[v]>d[u]+length[(u,v)]:
                d[v]=d[u]+length[(u,v)]
                dh[v]=d[v]+h[v]
                p[v]=u
                pq.append([v,dh[v]])

        if u==g:
            flag=True
            return [d[u],u,flag]
    else:
        flag=False
        return [flag]

x=search(graph,"Arad","Bucharest",flag=False)
if len(x)>1:
    pathcost=x[0]
    goal=x[1]
    flag=x[2]
else:
    flag=x[0]
if flag==False:
    print("NO PATH FOUND")
else:
    path=[]
    path.append(goal)
    while p[goal]!=None:
        path.append(p[goal])
        goal=p[goal]

    print("Path:",end="")
    for i in range(len(path)-1,-1,-1):
        if i==0:
            print(str(path[i]),end="")
        else:
            print(str(path[i])+"->",end="")
    print()
    print("Total distance:"+str(pathcost)+"km")

f1.close()


```
