```python
import random
import math
arr=[]
id=input()
x=int(id[4])
y=math.ceil(int(id[-1]+id[-2])*1.5)

class Tree:
    def __init__(self, elem):
        self.elem = elem
        self.position = []

def alpha_beta_pruning(node, depth, alpha, beta, max_player):
    if depth == 0 or not node.position:
        return node.elem

    if max_player==True:
        maxEval = -math.inf
        for child in node.position:
            eval=alpha_beta_pruning(child, depth - 1, alpha, beta, False)
            maxEval=max(maxEval,eval)
            alpha = max(alpha, eval)
            if beta <= alpha:
                break
        return maxEval
    else:
        minEval = math.inf
        for child in node.position:
            eval=alpha_beta_pruning(child, depth - 1, alpha, beta, True)
            minEval=min(minEval,eval)
            beta = min(beta, eval)
            if beta <= alpha:
                break
        return minEval

for i in range(8):
    arr.append(random.randint(x,y))

print("Generated 8 random points between the minimum and maximum point limits:",arr)

def creation():
    count=0

    max_1 = Tree(None)
    for i in range(2):
        min_1 = Tree(None)
        max_1.position.append(min_1)
        for j in range(2):
            max_2 = Tree(None)
            min_1.position.append(max_2)
            for k in range(2):
                min_2 = Tree(arr[count])
                count+=1
                max_2.position.append(min_2)
    return max_1


depth=3
root=creation()
limit=int(id[-1]+id[-2])
value=alpha_beta_pruning(root, depth, -math.inf, math.inf, True)

print("Total points to win:",limit)
print("Achieved point by applying alpha-beta pruning =",value)
if value>=limit:
    print("The winner is Optimus Prime")
else:
    print("The Winner is Megatron")


n=int(id[3])
shuffle=[]
count=0
if n==0:
    n=8
    for i in range(n):
        random.shuffle(arr)
        root=creation()
        depth=4
        limit=int(id[-1]+id[-2])
        value=alpha_beta_pruning(root, depth, -math.inf, math.inf, True)
        shuffle.append(value)
        if value>=limit:
            count+=1
else:
    for i in range(n):
        random.shuffle(arr)
        root=creation()
        depth=4
        limit=int(id[-1]+id[-2])
        value=alpha_beta_pruning(root, depth, -math.inf, math.inf, True)
        shuffle.append(value)
        if value>=limit:
            count+=1
print("After the shuffle:")
print("List of all points values from each shuffles:",shuffle)
print("The maximum value of all shuffles:",max(shuffle))
print("Won",count,"times out of",n,"number of shuffles")
```
